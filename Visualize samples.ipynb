```python
from dataset import CcmDataset
import matplotlib.pyplot as plt
```

```python
noise = [[0, 0], [0, 0.05], [0, 0.1], [0, 0.2]] #[mean, std] if mean == None, calculates mean automatically. 
fig, ax = plt.subplots(4, 1, figsize = (10,20))
for i, (mean, std) in enumerate(noise):
    data = CcmDataset("data/validation.pkl", mean, std)
    ax[i].imshow(data[1][0][0].permute(1,2,0))
    ax[i].set_title(f"mu : {mean}, std : {std}")
```

```python
blur = [[None, None], [3, 5], [5, 5], [7, 5]] #[kernel size, std] if None, no blurring. 
fig, ax = plt.subplots(4, 1, figsize = (10,20))
for i, (k, std) in enumerate(blur):
    data = CcmDataset("data/validation.pkl", 0, 0, k, std)
    ax[i].imshow(data[1][0][0].permute(1,2,0))
    ax[i].set_title(f"kernel size : {k}, std : {std}")
```
